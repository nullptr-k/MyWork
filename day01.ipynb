## 加载库目录

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 定义函数

```python
def p(x): 
    return (x-2)**9
def q(x):
    return x**9-18*x**8+144*x**7-672*x**6+2016*x**5-4032*x**4+5376*x**3-4608*x**2+2304*x-512
```

## 取区间[1.92，2.08]的值 分成1000份

```python
x=np.linspace(1.92,2.08,1000)
print(x)
```

## x作参数带入p函数

```python
y=p(x)
print(y)
```

## 画函数图

```python
plt.plot(x,y)
```

## x作参数带入函数q

```python
z=q(x)
print(z)
```

## 画函数图

```python
plt.plot(x,z)
```
